```python
# You are only allowed to use  the following libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_size = 100
```

```python
X1 = np.random.normal(5, 2, data_size).reshape(-1,1)
X2 = np.random.normal(10, 2, data_size).reshape(-1,1)
```

```python
w1_fixed = 4.8
w2_fixed = 2.5
w3_fixed = 1.2
```

```python
Y = w3_fixed*(w1_fixed*X1 + w2_fixed*X2)
```

```python
Y
```

```python
def my_ANN_model(x1, x2, y, Learning_Rate=0.01, nr_epochs=200, w1=-0.22, w2=0.17, w3=0.97, costs=[]):
    for epoch in range(nr_epochs):

        # Forward pass
        h1 = w1 * x1 + w2 * x2
        y_pred = w3 * h1
        
        # Calculate cost
        cost = np.mean((y_pred - y) ** 2)
        costs.append(cost)
        
        # Backward pass (gradient descent)
        dcost_dy_pred = 2 * (y_pred - y) / len(y)
        
        # Chain rule to compute gradients
        dy_pred_dw3 = np.mean(dcost_dy_pred * h1)
        dy_pred_h1 = dcost_dy_pred * w3

        dy_pred_dw1 = np.mean(dy_pred_h1 * x1)
        dy_pred_dw2 = np.mean(dy_pred_h1 * x2)
        
        # Update weights
        w3 -= Learning_Rate * dy_pred_dw3
        w1 -= Learning_Rate * dy_pred_dw1
        w2 -= Learning_Rate * dy_pred_dw2

        print(f"Epoch {epoch}, Cost: {cost:.6f}, Weights: w1={w1:.6f}, w2={w2:.6f}, w3={w3:.6f}")
    return w1, w2, w3, costs # weights associated with the two input parameters and a list containing the cost values for each training epoch
```

```python
w1, w2, w3, costs = my_ANN_model(X1, X2, Y)
```
